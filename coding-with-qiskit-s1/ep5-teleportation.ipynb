# Quantum Teleportation Algorithm
Teleports the state of $q0$ to $q2$.<br>
After execution, the state of $q0$ will collapsed during measurement.<br>
This is expected due to the no-cloning theorem.<br>

```python
from qiskit import *
```

```python
circuit = QuantumCircuit(3, 3)

%matplotlib inline
circuit.draw(output='mpl')
```

```python
circuit.x(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.h(1)
circuit.cx(1, 2)
circuit.draw(output='mpl')
```

```python
circuit.cx(0, 1)
circuit.h(0)
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.measure([0, 1], [0, 1])
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.cx(1, 2)
circuit.cz(0, 2)
circuit.draw(output='mpl')
```

```python
# we need to add "measure q2" to show result properly
# link: https://quantumcomputing.stackexchange.com/a/9028
circuit.measure(2, 2)

simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots=1024).result()
counts = result.get_counts()

from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
