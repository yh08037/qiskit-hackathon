# Fourier Checking Ciruit
Fourier checking circuit is an important tool used in signal processing domain.<br>
It is a fundamental circuit used in many other quantum algorithms for example, shor's algorithm.<br>


```python
import qiskit.quantum_info as qi
from qiskit.circuit.library import FourierChecking
from qiskit.visualization import plot_histogram
```

```python
f = [1, -1, -1, -1]
g = [1,  1, -1, -1]
```

```python
circ = FourierChecking(f=f, g=g)
circ.draw()
```

```python
zero = qi.Statevector.from_label('00')
sv = zero.evolve(circ)
probs = sv.probabilities_dict()
plot_histogram(probs)
```
